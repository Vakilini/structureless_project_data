```python
import pandas as pd
import numpy as np
import scipy.stats as stats
df = pd.read_excel('Stens.xlsx', header = (0,1))

df2 = pd.read_excel('Answer.xlsx',header=0,)
def pearsonr(x, y):
  # Assume len(x) == len(y)
  n = len(x)
  sum_x = float(sum(x))
  sum_y = float(sum(y))
  sum_x_sq = sum(xi*xi for xi in x)
  sum_y_sq = sum(yi*yi for yi in y)
  psum = sum(xi*yi for xi, yi in zip(x, y))
  num = psum - (sum_x * sum_y/n)
  den = pow((sum_x_sq - pow(sum_x, 2) / n) * (sum_y_sq - pow(sum_y, 2) / n), 0.5)
  if den == 0: return 0
  return num / den
df2
```

```python
f1 = list()
for a in df.columns:
    #print('a ',a)
    for b in df.columns:
        
        if a == b:
            continue
        else:
            d = (str(a)+" и "+ str(b))
            f1.append(d)

for a in range(len(df.columns) ):
    
    for b in range(len(df.columns) ):
        
        if a == b:
            continue
        else:
            r= pearsonr(df[df.columns[a]], df[df.columns[b]])
            if r < 0:
                t = stats.ttest_ind(df[df.columns[b]],df[df.columns[a]], equal_var=True)
                d = t[1]/2
                e = 'зависимости нет'
                l= 'не благоприятна'
                if d >= 0.5:
                    f = 'достоверный'
                else:
                    f = 'не достоверный'
            else:
                t = stats.ttest_rel(df[df.columns[b]],df[df.columns[a]], alternative="less")
                d = t[1]
                e = 'низкая зависимость'
                l= 'благоприятна'
                if d >= 0.5:
                    f = 'не достоверный'
                else:
                    f = 'достоверный'
           # t-test, p = s.ttest_1samp(df[df.columns[a]], df[df.columns[b]])
            #print(f1[len(df2.columns) - 1])
            df2[f1[len(df2.columns) - 1]] =  [r]*1 +[e]*1+[d]*1+[None]*2 +[f] + [r]*1 +[l]*1
            print(d)
df2   
```
